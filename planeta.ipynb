Hay que resolver $$\frac{d^2 \vec \rho}{d \tau²} = -\frac{\hat \rho}{\rho²}$$

```python
def mov_planeta(x,t):
    r3 = (x[0]**2 + x[1]**2)**(1.5)
    drx = x[2]
    dry = x[3]
    dvrx = -x[0]/r3
    dvry = -x[1]/r3
    return array([drx,dry,dvrx,dvry])

def J_mov_planeta(x,t):
    d_ec = array([0,0,0,0])
    r5 = (x[0]**2 + x[1]**2)**(2.5)
    rxmry = (2*x[0]*x[0]-x[1]*x[1])/r5
    rymrx = (2*x[1]*x[1]-x[0]*x[0])/r5
    rxry = (3*x[0]*x[1])/r5
    jac = array([0,0,1,0],[0,0,0,1],[rxmry,rxry,0,0],[rxry,rymrx,0,0])
    return d_ec,jac
```

```python
t,r_vr = rk2(mov_planeta,[1,0,0,1],[0,40000*pi],0.1)
plot(r_vr[:,0],r_vr[:,1])
show()
```
