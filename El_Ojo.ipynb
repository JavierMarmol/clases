```python
from pylab import *
%run "mis_int.ipynb"
%matplotlib inline
```

```python
def osc0(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])+0.1])
```

```python
def osc(y,t,w02=1.):
    return array([y[1],-w02*sin(y[0])+0.1*sin(t)])
```

```python
for i in range(1000):
    y0,u0 = 1.2*pi*(2*rand()-1),4*(2*rand()-1)
    t,sol=rk4(osc,[y0,u0],[0,10.])
    plot(sol[:,0],sol[:,1])
zoom = 4
xlim([-zoom,zoom]),ylim([-zoom,zoom])
axes().set_aspect('equal'),grid(True)
show()
```
